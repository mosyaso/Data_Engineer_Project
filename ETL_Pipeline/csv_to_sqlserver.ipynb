```python
#List of Library
import pyodbc
import pandas as pd
from datetime import datetime
```

```python
# Extract data from csv into dataframe
file_path = (r"C:\Users\syakir1937\Downloads\employees.csv")
df = pd.read_csv(file_path)

# Transform data
# Remove (.) in 'PHONE_NUMBER'
df['PHONE_NUMBER'] = df['PHONE_NUMBER'].str.replace(".", "")

# Change 'EMAIL' to lower case
df['EMAIL'] = df['EMAIL'].str.lower()

# Convert the 'Date' column to datetime objects
df['HIRE_DATE'] = pd.to_datetime(df['HIRE_DATE'], format='%d-%b-%y')

# Change the datetime format to ddmmyyyy
df['HIRE_DATE'] = df['HIRE_DATE'].dt.strftime('%d%m%Y')

display(df)
```

```python
# set up the connection parameters
server = 'ServerName'
database = 'Database'
username = 'Username'
password = 'Password'

# create a connection string
connection_string = f"DRIVER={{SQL Server}};SERVER={server};DATABASE={database};UID={username};PWD={password}"

# connect to the database
conn = pyodbc.connect(connection_string)
# create a cursor object
cursor = conn.cursor()
```

```python
table_name = 'Employees_Details'
# SQL query to drop table if it exists
drop_table = f"IF OBJECT_ID('{table_name}', 'U') IS NOT NULL DROP TABLE {table_name}"
								
cursor.execute(drop_table)

# Create Table
cursor.execute('''
        CREATE TABLE {}(
                EMPLOYEE_ID integer primary key,
                FIRST_NAME varchar(20),
                LAST_NAME varchar(20),
                EMAIL varchar(25),
                PHONE_NUMBER varchar(20),
                HIRE_DATE varchar(20),
                JOB_ID varchar(20),
                SALARY decimal(10,2),
                COMMISSION_PCT varchar(5),
                MANAGER_ID integer,
                DEPARTMENT_ID integer
                )
            '''.format(table_name))
```

```python
#Insert Into Table						
for row in df.itertuples(index=False):
   cursor.execute('''
                INSERT INTO dbo.{} (EMPLOYEE_ID, FIRST_NAME, LAST_NAME, EMAIL, PHONE_NUMBER, HIRE_DATE, JOB_ID, SALARY, COMMISSION_PCT, MANAGER_ID, DEPARTMENT_ID)
                values(?,?,?,?,?,?,?,?,?,?,?)
                '''.format(table_name),
                row.EMPLOYEE_ID, row.FIRST_NAME, row.LAST_NAME, row.EMAIL, row.PHONE_NUMBER, row.HIRE_DATE, row.JOB_ID, row.SALARY, row.COMMISSION_PCT, row.MANAGER_ID, row.DEPARTMENT_ID
                )
conn.commit()

cursor.close()
```
